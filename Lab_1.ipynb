## Lists

Perhaps the greatest strength of python is iterating over lists.  Use the difference formula of calculus to show that the derivative of $\ln x$ at 1 is 1.

In other words show that as $h$ gets smaller and smaller $\frac{\ln(1+h)-\ln(1)}h$ approaches 1.  Mathematically we write this as
$$
\lim_{h\to 0} \frac{\ln(1+h)-\ln(1)}h = 1
$$


Use the list provided below.

Extra challenge:  Use the `sympy` package to compute the limit and derivative.

```python
import math

h = [1,0.1,0.001,0.002,0.0001,0.0000000001]

for i in h:
    limit_result = (math.log(1+i) - math.log(1))/i
    print (" For h = ", i, " the result is: ", limit_result)
```

The above result shows that as h gets closer to zero, the value of the limit goes closer to 1. This shows that the derivative of $\ln x$ at 1 is 1

### Extra Challenge
Using sympy to compute the limit and derivative

```python
import sympy as sp

h = sp.symbols("h")

#Difference quotient at x = 1
difference_quotient = (sp.log(1 + h) - sp.log(1))/h

#limit as h tends to 0
derivative = sp.limit(difference_quotient, h, 0)

print("Using sympy, limit of ln(x) as x -> 1: ", derivative)
```

```python
import sympy as sp

x = sp.symbols("x")

f = sp.log(x)

derivative_function = sp.diff(f, x)
value_at_1 = derivative_function.subs(x, 1)

print("Derivative of ln(x) using sympy: ", derivative_function)
print("Derivative of ln(x) at 1: ", value_at_1)
```

## Truth of Summation

Verify the following summation formula using python for all $n$ where $n\in\mathbb{Z}$ and $n\leq 1000$.

$$
\sum_{i = 0}^n i^3 = \left(\frac{n(n+1)}{2}\right)^2
$$

Extra challenge:  Reduce the output to a single true or false.  Do not attempt until original has been verified with a clean code.

```python
for n in range(1001):
    left_value = sum(i**3 for i in range(n+1))
    right_value = (n*(n+1)//2)**2
    print ("When i = ", n, ", LHS = ", left_value, ", RHS = ", right_value, ". LHS = RHS is ", left_value==right_value)
```

### Extra Challenge
Returning a single true or false, depending on whether the summation holds true for all numbers in the range

```python
def is_summation_correct(max_n):
    for n in range(max_n + 1):
        left_value = sum(i**3 for i in range(n+1))
        right_value = (n*(n+1)//2)**2
        if not (left_value == right_value):
            return False
    return True

print(is_summation_correct(1000))
```
